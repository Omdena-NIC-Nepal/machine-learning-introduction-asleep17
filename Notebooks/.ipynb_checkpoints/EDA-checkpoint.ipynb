<span>Visualize relationships between features and the target variable.</span>

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Correlatoin Heatmap between features and the target variable
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap with MEDV')
plt.show()
```

```python
#scatterplot between Rooms and the Median value of the house
sns.scatterplot(data=df, x='rm', y='medv')
plt.title('ROOM vs MEDV (Housing Prices)')
plt.xlabel('Rooms in the house (RM)')
plt.ylabel('Median Value of Homes (MEDV)')
plt.show()
```

```python
#scatterplot between Rooms and the Median value of the house
sns.scatterplot(data=df, x='lstat', y='medv')
plt.title('LSTAT vs MEDV (Housing Prices)')
plt.xlabel('Percentage of Lower Status Population (LSTAT)')
plt.ylabel('Median Value of Homes (MEDV)')
plt.show()
```

```python
sns.scatterplot(data=df, x='tax', y='medv')
plt.title('Tax vs MEDV (Housing Prices)')
plt.xlabel('Tax rate per 10,000$ (RM)')
plt.ylabel('Median Value of Homes (MEDV)')
plt.show()
```
